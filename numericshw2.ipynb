```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, np.pi, 31)
y = np.cos(t)
S = np.sum(t * y)
print(f"The sum is: {S}")
```

Part **b** of 4

```python
R = 1.2
deltar = 0.1
f = 15
p = 0
```

```python
#one curve
theta = np.linspace(0, 2*np.pi, 400)
x = R*(1+deltar*np.sin(f*theta+p))*np.cos(theta)
y = R*(1+deltar*np.sin(f*theta+p))*np.sin(theta)

plt.figure()
plt.plot(x, y)
plt.title("Wavy Circle no for loop")
plt.show()
```

```python
#second curve
for i in range(1, 11):
    R  = i
    deltar = 0.05
    f = 2+i
    p = np.random.uniform(0, 2)
    x = R*(1+deltar*np.sin(f*theta+p))*np.cos(theta)
    y = R*(1+deltar*np.sin(f*theta+p))*np.sin(theta)

    plt.plot(x, y, label=f"i={i}")

plt.title("10 diff with for loop")
plt.legend()
plt.show()
```
