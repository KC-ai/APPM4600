```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pade33(x):
    return (x - (7/60)*x**3) / (1 + (1/20)*x**2)
```

```python
def pade24(x):
    return x / (1 + (1/6)*x**2 + (7/360)*x**4)
```

```python
def pade42(x):
    return (x - (7/60)*x**3) / (1 + (1/20)*x**2)
```

```python
def maclaurin(x):
    return x - (x**3)/6 + (x**5)/120
```

```python
x = np.linspace(0, 5, 1000)
sinx = np.sin(x)
```

```python
error_p33 = np.abs(sinx - pade33(x))
error_p24 = np.abs(sinx - pade24(x))
error_p42 = np.abs(sinx - pade42(x))
error_mac  = np.abs(sinx - maclaurin(x))

plt.figure(figsize=(10, 6))
plt.plot(x, error_p33, label='Pade 3/3 and 4/2)')
plt.plot(x, error_p24, label='Pade 2/4')
plt.plot(x, error_mac,  label='6th order Maclaurin')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Error Comparisons')
plt.legend()
plt.grid(True)
plt.show()
```
